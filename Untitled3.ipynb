```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import MinMaxScaler

# Завдання 1: Згенерувати вибадковий набір даних в діапазоні 1000 значень
def generate_data(num_samples=1000):
    np.random.seed(25)  # Встановлюємо випадкове насіння для відтворюваності
    X = np.random.uniform(low=50, high=200, size=num_samples)  # Рівномірний розподіл
    Y = 3 * X + np.random.randn(num_samples) * 80 + 150  # Зміщення для Y
    return X, Y

X, Y = generate_data()  # Використовуємо 1000 згенерованих значень

# Візуалізація згенерованих даних
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='cyan', alpha=0.8)
plt.title('Розподіл згенерованих даних (рівномірний розподіл)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

# Завдання 2: Нормалізувати значення
def normalize_data(X):
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X.reshape(-1, 1))  # Нормалізація в діапазоні [0, 1]
    return X_scaled

X_scaled = normalize_data(X)  # Нормалізуємо X

# Завдання 3: Розділити існуючі записи на навчальну та тестову вибірки
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.4, random_state=21)

# Візуалізація навчальної та тестової вибірки
plt.figure(figsize=(10, 6))
plt.scatter(X_train, Y_train, color='magenta', alpha=0.5, label='Навчальна вибірка')
plt.scatter(X_test, Y_test, color='lime', alpha=0.5, label='Тестова вибірка')
plt.title('Поділ на навчальну та тестову вибірки')
plt.xlabel('X (нормалізоване)')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Завдання 4: Навчити KNN-регресор з різними значеннями K
def train_knn(X_train, Y_train, X_test, optimal_k):
    knn = KNeighborsRegressor(n_neighbors=optimal_k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    return Y_pred

# Пошук оптимального K (параметр для вибору K залишився, але без MSE)
k_values = range(1, 21)  # Перевіряємо K від 1 до 20
Y_pred_best = None

# Навчання моделі KNN з різними значеннями K і вибір одного з них
optimal_k = 15  # Вибрано K без використання MSE
Y_pred_best = train_knn(X_train, Y_train, X_test, optimal_k)

# Завдання 5: Візуалізація прогнозування за допомогою KNN (K=оптимальне)
plt.figure(figsize=(10, 6))
plt.scatter(X_test, Y_test, color='purple', alpha=0.5, label='Тестові значення Y')
plt.scatter(X_test, Y_pred_best, color='orange', alpha=0.5, label='Прогнозовані значення Y')
plt.title(f'Прогнозування за допомогою KNN (K={optimal_k})')
plt.xlabel('X (нормалізоване)')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
